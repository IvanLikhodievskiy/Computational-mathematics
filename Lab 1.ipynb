# Лабораторная работа 1

### Функции:

- $\sin{(x^2)}$
- $\cos{(\sin{(x)})}$
- $\exp{(\sin{(\cos{(x)})})}$
- $\ln{(x + 3)}$
- $(x + 3)^{0.5}$

### Производные рассматриваемых функций

- $\left(\sin{(x^2)}\right)' = 2x \cdot \cos{(x^2)}$
- $\left(\cos{(\sin{(x)})}\right)' = -\sin{(\sin{x})} \cdot \cos{x}$
- $\left(\exp{(\sin{(\cos{(x)})})}\right)' = -e^{\sin{(\cos{x})}} \cdot \cos{(\cos{x})} \cdot \sin{x}$
- $\left(\ln{(x + 3)}\right)' = \frac{1}{x + 3}$
- $\left((x + 3)^{0.5}\right)' = \frac{1}{2\sqrt{x + 3}}$

### Формулы для численного расчёта производных:

1. $\frac{df}{dx}(x) = \frac{f(x+h) - f(x)}{h}$

2. $\frac{df}{dx}(x) = \frac{f(x) - f(x-h)}{h}$

3. $\frac{df}{dx}(x) = \frac{f(x+h) - f(x-h)}{2h}$

4. $\frac{df}{dx}(x) = \frac{4}{3}\frac{f(x+h) - f(x-h)}{2h} - \frac{1}{3}\frac{f(x+2h) - f(x-2h)}{4h}$

5. $\frac{df}{dx}(x) = \frac{3}{2}\frac{f(x+h) - f(x-h)}{2h} - \frac{3}{5}\frac{f(x+2h) - f(x-2h)}{4h} + \frac{1}{10}\frac{f(x+3h) - f(x-3h)}{6h}$

Считаем производные в точке $x = 7$


```python
class Function:
    def __init__(self, func, deriv, name):
        self.func = func
        self.deriv = deriv
        self.name = name

class Num_Deriv:
    def __init__(self, func, name):
        self.func = func
        self.name = name
```

```python
# First function
def f1(x):
    return np.sin(x ** 2)

# Second function
def f2(x):
    return np.cos(np.sin(x))

# Third function
def f3(x):
    return np.exp(np.sin(np.cos(x)))

# Fourth function
def f4(x):
    return np.log(x + 3)

# Fifth function
def f5(x):
    return np.sqrt(x + 3)
```

```python
# Analytical derivative of f1
def deriv_f1(x):
    return 2 * x * np.cos(x ** 2)

# Analytical derivative of f2
def deriv_f2(x):
    return -np.sin(np.sin(x)) * np.cos(x)

# Analytical derivative of f3
def deriv_f3(x):
    return -np.sin(x) * np.cos(np.cos(x)) * np.exp(np.sin(np.cos(x)))

# Analytical derivative of f4
def deriv_f4(x):
    return 1 / (x + 3)

# Analytical derivative of f5
def deriv_f5(x):
    return 1 / (2 * np.sqrt(x + 3))
```

```python
# Numerical derivative by formula 1
def num_deriv_1(f, x, h):
    return (f(x + h) - f(x)) / h

# Numerical derivative by formula 2
def num_deriv_2(f, x, h):
    return (f(x) - f(x - h)) / h

# Numerical derivative by formula 3
def num_deriv_3(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)

# Numerical derivative by formula 4
def num_deriv_4(f, x, h):
    return (4 / 3) * ((f(x + h) - f(x - h)) / (2 * h)) - (1 / 3) * ((f(x + 2 * h) - f(x - 2 * h)) / (4 * h))

# Numerical derivative by formula 5
def num_deriv_5(f, x, h):
    return (3 / 2) * ((f(x + h) - f(x - h)) / (2 * h)) - (3 / 5) * ((f(x + 2 * h) - f(x - 2 * h)) / (4 * h)) + (1 / 10) * ((f(x + 3 * h) - f(x - 3 * h)) / (6 * h))
```

```python
x = 7.0

h = [(2 / 2 ** i) for i in range(1, 22)]

functions = [
                 Function(f1, deriv_f1, "$\\sin{(x^2)}$"),
                 Function(f2, deriv_f2, "$\\cos{(\\sin{(x)})}$"),
                 Function(f3, deriv_f3, "$\\exp{(\\sin{(\\cos{(x)})})}$"),
                 Function(f4, deriv_f4, "$\\ln{(x + 3)}$"),
                 Function(f5, deriv_f5, "$(x + 3)^{0.5}$"),
            ]

num_derivs = [
                  Num_Deriv(num_deriv_1, "$\\frac{df}{dx}(x) = \\frac{f(x+h) - f(x)}{h}$"),
                  Num_Deriv(num_deriv_2, "$\\frac{df}{dx}(x) = \\frac{f(x) - f(x-h)}{h}$"),
                  Num_Deriv(num_deriv_3, "$\\frac{df}{dx}(x) = \\frac{f(x+h) - f(x-h)}{2h}$"),
                  Num_Deriv(num_deriv_4, "$\\frac{df}{dx}(x) = \\frac{4}{3}\\frac{f(x+h) - f(x-h)}{2h} - \\frac{1}{3}\\frac{f(x+2h) - f(x-2h)}{4h}$"),
                  Num_Deriv(num_deriv_5, "$\\frac{df}{dx}(x) = \\frac{3}{2}\\frac{f(x+h) - f(x-h)}{2h} - \\frac{3}{5}\\frac{f(x+2h) - f(x-2h)}{4h} + \\frac{1}{10}\\frac{f(x+3h) - f(x-3h)}{6h}$"),
             ]

func_1 = [[], [], [], [], []]
func_2 = [[], [], [], [], []]
func_3 = [[], [], [], [], []]
func_4 = [[], [], [], [], []]
func_5 = [[], [], [], [], []]

data = [func_1, func_2, func_3, func_4, func_5]

for arr, f in zip(data, functions):
    for func_arr, num_deriv in zip(arr, num_derivs):
        for step in h:
            func_arr.append(np.fabs(num_deriv.func(f.func, x, step) - f.deriv(x)))
```

```python
for arr, f in zip(data, functions):
    plt.figure(figsize=[14, 7])

    plt.title(f.name, fontsize=30)
    
    plt.xscale("log")
    plt.yscale("log")

    plt.xlabel('$\\log{h}$', fontsize = 20)
    plt.xticks (fontsize = 20, ha = "center", va = "top")

    plt.ylabel("$\\log{(\\text{Numerical derivative error})}$", fontsize = 20)
    plt.yticks (fontsize = 20, rotation = 30, ha = "right", va = "top")

    plt.grid(which = "both")
    for func_arr, num_deriv in zip(arr, num_derivs):
        plt.scatter(np.array(h), np.array(func_arr), label=num_deriv.name)
        plt.plot(np.array(h), np.array(func_arr))
        plt.legend (loc = "best")
        plt.tight_layout()
```
